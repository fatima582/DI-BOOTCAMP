```python
# Exercice3 : Ingénierie des fonctionnalités
import pandas as pd
df_processed = pd.read_csv('titanic dataset\\titanic dataset\\train.csv')
df_processed = df_processed.drop('Cabin', axis=1)
median_age = df_processed['Age'].median()
df_processed['Age'].fillna(median_age, inplace=True)
mode_embarked = df_processed['Embarked'].mode()[0]
df_processed['Embarked'].fillna(mode_embarked, inplace=True)

print("DataFrame nettoyé pour l'Exercice 3 :")
display(df_processed.head())
print("\nInformations sur le DataFrame nettoyé :")
df_processed.info()
df_processed['FamilySize'] = df_processed['SibSp'] + df_processed['Parch'] + 1

print("DataFrame avec la nouvelle colonne 'FamilySize' :")
display(df_processed[['SibSp', 'Parch', 'FamilySize']].head())

df_processed['Title'] = df_processed['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

df_processed['Title'] = df_processed['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
df_processed['Title'] = df_processed['Title'].replace('Mlle', 'Miss')
df_processed['Title'] = df_processed['Title'].replace('Ms', 'Miss')
df_processed['Title'] = df_processed['Title'].replace('Mme', 'Mrs')

print("Répartition des titres extraits et regroupés :")
display(df_processed['Title'].value_counts())

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df_processed['Sex_encoded'] = le.fit_transform(df_processed['Sex'])
print("Encodage de 'Sex' :")
print(df_processed[['Sex', 'Sex_encoded']].drop_duplicates().sort_values('Sex_encoded'))

df_processed['Embarked_encoded'] = le.fit_transform(df_processed['Embarked'])
print("\nEncodage de 'Embarked' :")
print(df_processed[['Embarked', 'Embarked_encoded']].drop_duplicates().sort_values('Embarked_encoded'))

df_processed['Title_encoded'] = le.fit_transform(df_processed['Title'])
print("\nEncodage de 'Title' :")
print(df_processed[['Title', 'Title_encoded']].drop_duplicates().sort_values('Title_encoded'))

print("\nDataFrame avec les colonnes encodées (premières 5 lignes) :")
display(df_processed.head())
```

```python
# Exercice4 : Détection et traitement des valeurs aberrantes

import matplotlib.pyplot as plt
import seaborn as sns

df_outliers = df_processed.copy()

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(y=df_outliers['Age'])
plt.title('Boxplot de l\'Age')

plt.subplot(1, 2, 2)
sns.histplot(df_outliers['Age'], kde=True)
plt.title('Distribution de l\'Age')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(y=df_outliers['Fare'])
plt.title('Boxplot du Tarif')

plt.subplot(1, 2, 2)
sns.histplot(df_outliers['Fare'], kde=True)
plt.title('Distribution du Tarif')
plt.tight_layout()
plt.show()

def detect_iqr_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Valeurs aberrantes détectées dans '{column}' (basé sur IQR):")
    if not outliers.empty:
        display(outliers[[column]].head())
        print(f"Nombre de valeurs aberrantes dans '{column}': {len(outliers)}")
    else:
        print("Aucune valeur aberrante détectée.")
    return lower_bound, upper_bound

print("Détection des valeurs aberrantes pour 'Age' :")
lower_age, upper_age = detect_iqr_outliers(df_outliers, 'Age')

print("\nDétection des valeurs aberrantes pour 'Fare' :")
lower_fare, upper_fare = detect_iqr_outliers(df_outliers, 'Fare') 
df_capping = df_outliers.copy()

upper_fare_cap = df_capping['Fare'].quantile(0.98)

df_capping['Fare_capped'] = df_capping['Fare'].apply(lambda x: upper_fare_cap if x > upper_fare_cap else x)
print(f"Valeur maximale de 'Fare' avant plafonnement : {df_capping['Fare'].max():.2f}")
print(f"Seuil de plafonnement (98ème percentile) : {upper_fare_cap:.2f}")
print(f"Valeur maximale de 'Fare_capped' après plafonnement : {df_capping['Fare_capped'].max():.2f}")

plt.figure(figsize=(10, 5))
sns.histplot(df_capping['Fare_capped'], kde=True)
plt.title('Distribution du Tarif après Plafonnement (98e percentile)')
plt.show()

import numpy as np
df_log_transform = df_outliers.copy()
df_log_transform['Fare_log'] = np.log1p(df_log_transform['Fare'])

print("Distribution de 'Fare' avant transformation logarithmique :")
sns.histplot(df_log_transform['Fare'], kde=True)
plt.title('Distribution Originale du Tarif')
plt.show()

print("\nDistribution de 'Fare' après transformation logarithmique :")
sns.histplot(df_log_transform['Fare_log'], kde=True)
plt.title('Distribution du Tarif après Transformation Logarithmique (log1p)')
plt.show()
```

```python
# Exercice6 : Encodage des caractéristiques

df_final_encoding = pd.read_csv('titanic.csv')

df_final_encoding = df_final_encoding.drop('Cabin', axis=1)

median_age = df_final_encoding['Age'].median()
df_final_encoding['Age'].fillna(median_age, inplace=True)

mode_embarked = df_final_encoding['Embarked'].mode()[0]
df_final_encoding['Embarked'].fillna(mode_embarked, inplace=True)

df_final_encoding['FamilySize'] = df_final_encoding['SibSp'] + df_final_encoding['Parch'] + 1
df_final_encoding['Title'] = df_final_encoding['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
df_final_encoding['Title'] = df_final_encoding['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
df_final_encoding['Title'] = df_final_encoding['Title'].replace('Mlle', 'Miss')
df_final_encoding['Title'] = df_final_encoding['Title'].replace('Ms', 'Miss')
df_final_encoding['Title'] = df_final_encoding['Title'].replace('Mme', 'Mrs')

print("DataFrame préparé pour l'encodage final :")
display(df_final_encoding.head())
print("\nInformations sur le DataFrame :")
df_final_encoding.info()

categorical_cols = ['Sex', 'Embarked', 'Title']
print("Colonnes catégorielles à encoder :", categorical_cols)

df_encoded = pd.get_dummies(df_final_encoding, columns=categorical_cols, drop_first=True)
print("DataFrame après encodage One-Hot (premières 5 lignes) :")
display(df_encoded.head())
print("\nInformations sur le DataFrame encodé :")
df_encoded.info()
```

```python
# Exercice7: Transformation des données pour la caractéristique Âge
df_age_grouped = df_encoded.copy()

age_bins = [0, 12, 18, 60, 100]
age_labels = ['Enfant', 'Adolescent', 'Adulte', 'Senior']
df_age_grouped['AgeGroup'] = pd.cut(df_age_grouped['Age'], bins=age_bins, labels=age_labels, right=False)

print("Répartition des passagers par groupe d'âge :")
display(df_age_grouped['AgeGroup'].value_counts())

print("\nAperçu du DataFrame avec la nouvelle colonne 'AgeGroup' :")
display(df_age_grouped[['Age', 'AgeGroup']].head())

df_final = pd.get_dummies(df_age_grouped, columns=['AgeGroup'], drop_first=True)
print("DataFrame après encodage One-Hot des groupes d'âge (premières 5 lignes) :")
display(df_final.head())
print("\nInformations sur le DataFrame final :")
df_final.info()
```
