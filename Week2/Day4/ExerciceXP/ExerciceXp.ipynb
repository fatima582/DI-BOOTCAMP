```python
#Exercice2
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [72, 74, 76, 80, 82, 78, 75]
plt.plot(x, y)
plt.show()

plt.plot(x, y, color='green', linestyle='dashed', linewidth=2, marker='o', markerfacecolor='blue', markersize=12)
plt.show()
```

```python
#Exercice3
import matplotlib.pyplot as plt
categories = ["Janvier", "Février", "Mars", "Avril", "Mai"]
values = [5000, 5500, 6200, 7000, 7500]
plt.bar(categories, values)
plt.title("Ventes mensuelles")
plt.xlabel("Mois")
plt.ylabel("Ventes")
plt.show()
```

```python
#Exercice4
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv("Student Mental health\\Student Mental health.csv")

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="What is your CGPA?", color="skyblue", kde=True)

plt.title(
    "Distribution des moyennes cumulatives (CGPA) des étudiants", fontsize=14
)
plt.xlabel("Moyenne Cumulative (CGPA)", fontsize=12)
plt.ylabel("Nombre d'étudiants", fontsize=12)

plt.show()
```

```python
#Exercice5

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

plt.figure(figsize=(7, 5))
sns.countplot(
    data=df,
    x="Choose your gender",
    hue="Do you have Anxiety?",
    palette="pastel",
)

plt.title(
    "Comparaison de la présence d'anxiété selon le sexe des étudiants",
    fontsize=14,
)
plt.xlabel("Sexe (Gender)", fontsize=12)
plt.ylabel("Nombre d'étudiants", fontsize=12)
plt.legend(title="Souffre d'anxiété ?")

plt.show()
```

```python
#Exercice6
import matplotlib.pyplot as plt
import seaborn as sns

df["Panic_numeric"] = df["Do you have Panic attack?"].map({"Yes": 1, "No": 0})

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x="Age",
    y="Panic_numeric",
    hue="Do you have Panic attack?",
    style="Do you have Panic attack?",
    s=100,
    palette="dark",
)

plt.title(
    "Relation entre l'âge et la survenue de crises de panique", fontsize=14
)
plt.xlabel("Âge des étudiants", fontsize=12)
plt.ylabel("Crise de panique (Non = 0 / Oui = 1)", fontsize=12)
plt.yticks([0, 1], ["0 (Non)", "1 (Oui)"])

plt.show()
```
