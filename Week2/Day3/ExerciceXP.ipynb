```python
#Exercice2
import numpy as np
from scipy import stats
data = [12, 15, 13, 12, 18, 20, 22, 21]
moyenne =np.mean(data)
mediane =np.median(data)
variance =np.var(data, ddof=0)
ecart_type =np.std(data, ddof=0)

resume = stats.describe(data)
print(f"Moyenne :{moyenne} ")
print (f"Médiane :{mediane}")
print (f"Variance :{variance}")
print (f"Écart-type :{ecart_type:.3f}")
print (resume)
```

```python
#Exercice3

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

moyenne =50
ecart_type =10
x = np.linspace(moyenne - 4 * ecart_type, moyenne + 4 * ecart_type, 1000)
pdf = norm.pdf(x, loc=moyenne, scale=ecart_type)
pdf = norm.pdf(x, loc=moyenne, scale=ecart_type)

plt.figure(figsize=(8, 5))
plt.plot(
    x, pdf, label="Distribution Normale", color="blue", linewidth=2
)
plt.axvline(
    moyenne,
    color="red",
    linestyle="--",
    linewidth=1.5,
    label=f"Moyenne ({moyenne})",
)
plt.axvline(
    moyenne - ecart_type,
    color="orange",
    linestyle=":",
    label="Moyenne ± Écart-type",
)
plt.axvline(moyenne + ecart_type, color="orange", linestyle=":")

plt.title("Distribution Normale ($\mu = 50$, $\sigma = 10$)", fontsize=14)
plt.xlabel("Valeurs de X", fontsize=12)
plt.ylabel("Densité de probabilité", fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()

```

```python
# Exercice4
import numpy as np
from scipy import stats

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)


t_statistic, p_value = stats.ttest_ind(data1, data2)

print(f"Statistique t: {t_statistic:.2f}")
print(f"Valeur p: {p_value:.3f}")

# Interpréter le résultat
alpha = 0.05
if p_value < alpha:
    print("Il existe une différence significative entre les moyennes des deux échantillons (rejeter H0).")
else:
    print("Il n'y a pas de différence significative entre les moyennes des deux échantillons (ne pas rejeter H0).")
```

```python
# Exercice5

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

house_sizes = np.array([50, 70, 80, 100, 120]).reshape(-1, 1) # Redimensionner pour sklearn
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

model = LinearRegression()

model.fit(house_sizes, house_prices)
pente = model.coef_[0]
ordonnee_origine = model.intercept_

print(f"Pente de la droite de régression (coefficient): {pente:.2f}")
print(f"Ordonnée à l'origine: {ordonnee_origine:.2f}")

house_size_90 = np.array([[90]])
predicted_price_90 = model.predict(house_size_90)[0]
print(f"Prix estimé d'une maison de 90 mètres carrés: {predicted_price_90:.2f}")

print("\nInterprétation de la pente:")
print(f"La pente de {pente:.2f} signifie que pour chaque mètre carré supplémentaire, le prix de la maison augmente d'environ {pente:.2f} unités monétaires.")

plt.figure(figsize=(10, 6))
plt.scatter(house_sizes, house_prices, color='blue', label='Données réelles')
plt.plot(house_sizes, model.predict(house_sizes), color='red', label='Droite de régression')
plt.scatter(house_size_90, predicted_price_90, color='green', marker='x', s=100, label=f'Prédiction pour 90 m² ({predicted_price_90:.2f})')
plt.title('Régression Linéaire: Prix des maisons vs Taille')
plt.xlabel('Taille de la maison (m²)')
plt.ylabel('Prix de la maison')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Exercice6
import numpy as np
from scipy import stats

fertilizer_1 = np.array([5, 6, 7, 6, 5])
fertilizer_2 = np.array([7, 8, 7, 9, 8])
fertilizer_3 = np.array([4, 5, 4, 3, 4])

f_statistic, p_value_anova = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print(f"Statistique F: {f_statistic:.2f}")
print(f"Valeur p (ANOVA): {p_value_anova:.3f}")

alpha_anova = 0.05
if p_value_anova < alpha_anova:
    print("\nSelon la valeur p, il existe une différence significative entre les moyennes de croissance des plantes pour les différents engrais. (Rejeter H0)")
else:
    print("\nSelon la valeur p, il n'y a pas de différence significative entre les moyennes de croissance des plantes pour les différents engrais. (Ne pas rejeter H0)")

if p_value_anova >= alpha_anova:
    print("\nSi la valeur P était supérieure à 0,05 (ou si elle est effectivement supérieure), cela signifierait que nous n'avons pas suffisamment de preuves statistiques pour conclure que les engrais ont des effets significativement différents sur la croissance des plantes. Nous ne pourrions pas rejeter l'hypothèse nulle (H0), qui stipule qu'il n'y a pas de différence significative entre les moyennes de croissance des groupes d'engrais. En d'autres termes, les différences observées dans la croissance des plantes entre les groupes seraient probablement dues au hasard, et non à l'effet des engrais eux-mêmes.")
```

```python
#Exercice8
import pandas as pd
from scipy.stats import pearsonr, spearmanr

data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})
pearson_corr, _ = pearsonr(data['age'], data['income'])
print(f"Coefficient de corrélation de Pearson: {pearson_corr:.2f}")
spearman_corr, _ = spearmanr(data['age'], data['income'])
print(f"Coefficient de corrélation de Spearman: {spearman_corr:.2f}")
```
