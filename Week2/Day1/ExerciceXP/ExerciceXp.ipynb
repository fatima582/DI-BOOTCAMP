```python
#Exercice
import pandas as pd

def analyser_dataset(chemin_fichier, nom_du_dataset):
    print(f"\n" + "="*60)
    print(f"  ANALYSE DU DATASET : {nom_du_dataset}")
    print("="*60 + "\n")
    
    try:
        df = pd.read_csv(chemin_fichier)
        
        print("--- 1. Aperçu des 5 premières lignes ---")
        display(df.head())
        print("\n" + "-"*50 + "\n")
        
        print("--- 2. Description initiale du dataset ---")
        nb_lignes, nb_colonnes = df.shape
        print(f"• Dimensions : {nb_lignes} lignes et {nb_colonnes} colonnes.")
        
        print("\n• Liste des colonnes et types de données :")
        print(df.dtypes)
        
        print("\n• Nombre de valeurs manquantes par colonne :")
        print(df.isnull().sum())
        
        print("\n• Statistiques descriptives de base :")
        display(df.describe(include='all'))
        
    except FileNotFoundError:
        print(f" Erreur : Le fichier est introuvable au chemin : '{chemin_fichier}'")
    except Exception as e:
        print(f" Une erreur inattendue est survenue : {e}")

# En démarrant directement par le nom du sous-dossier, Python trouvera le fichier sans problème

chemin_sommeil = r'How Much Sleep Do Americans\How Much Sleep Do Americans\Time Americans Spend Sleeping.csv' 
analyser_dataset(chemin_sommeil, "Combien de sommeil les Américains obtiennent-ils réellement ?")

chemin_sante_mentale = r'Global Trends in Mental Health Disorder\Global Trends in Mental Health Disorder\Mental health Depression disorder Data.csv' 
analyser_dataset(chemin_sante_mentale, "Tendances mondiales en matière de troubles de santé mentale")

chemin_credit = r'Credit Card Approvals (Clean Data)\Credit Card Approvals (Clean Data)\clean_dataset.csv' 
analyser_dataset(chemin_credit, "Approbations de cartes de crédit")
```

Donnée quantitative : Une valeur numérique mesurable sur laquelle les opérations mathématiques (comme la moyenne ou la somme) ont un sens réel.

Donnée qualitative : Une valeur textuelle ou un code numérique qui sert à classer les observations dans des catégories ou des groupes.



# 1. Dataset : Combien de sommeil les Américains obtiennent-ils réellement ?

| Nom de la colonne | Type de donnée | Raisonnement |

| `index` | **Qualitative** | C'est un simple identifiant de ligne. Faire la moyenne des index n'a aucun sens mathématique. |
| `Year` | **Qualitative** | C'est un repère temporel qui sert de catégorie pour situer l'année de l'enregistrement. |
| `Period` | **Qualitative** | Variable textuelle indiquant une période de l'année (catégorie). |
| `Avg hrs per day sleeping` | **Quantitative** | Valeur numérique continue (le nombre d'heures). On peut calculer une moyenne globale de sommeil. |
| `Standard Error` | **Quantitative** | C'est une mesure statistique numérique représentant l'écart-type ou l'erreur de mesure. |
| `Type of Days` | **Qualitative** | Texte indiquant la catégorie de jour (ex: Semaine ou Week-end). |
| `Age Group` | **Qualitative** | Tranches d'âge sous forme de texte (ex: "18-24"). Ce sont des catégories. |
| `Activity` | **Qualitative** | Chaîne de caractères décrivant l'activité mesurée (le sommeil). |
| `Sex` | **Qualitative** | Variable catégorielle représentant le genre (Homme / Femme). |

Exercice6

la différence réside dans l'organisation : les données structurées suivent un modèle ou un tableau strict (lignes, colonnes, types définis), tandis que les données non structurées n'ont pas de format prédéfini et sont plus brutes

Les rapports financiers d'une entreprise sont stockés dans un fichier Excel.
Réponse : Données structurées (ou semi-structurées selon la complexité du fichier, mais généralement classées en structurées).
Pourquoi ? Les données sont organisées en lignes et en colonnes bien définies (chiffres d'affaires, dates, dépenses) permettant des calculs automatiques.


Photographies téléchargées sur une plateforme de médias sociaux.
Réponse : Données non structurées.
Pourquoi ? Une image est un ensemble de pixels. Même s'il y a des métadonnées cachées (date, lieu), le contenu de l'image en lui-même n'est pas rangé dans un tableau.


Une collection d'articles de presse sur un site web.
Réponse : Données non structurées.
Pourquoi ? Il s'agit de texte libre (du langage naturel). Bien qu'il y ait des titres et des paragraphes, le cœur de l'information n'obéit pas à une structure de base de données rigide.


Données d'inventaire dans une base de données relationnelle.
Réponse : Données structurées.
Pourquoi ? C'est l'exemple parfait ! Une base de données relationnelle (SQL) impose un schéma strict avec des tables, des clés primaires et des types de données précis 


Enregistrements d'entretiens issus d'une étude de marché.
Réponse : Données non structurées.
Pourquoi ? Qu'il s'agisse de fichiers audio ou de transcriptions textuelles, les discussions et les avis sont des données brutes qui nécessitent une analyse humaine ou une IA pour en extraire des tendances
